```python
#Question 1
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data/crypto.csv')
plt.rcParams['figure.figsize'] = [15, 9]
fig,ax = plt.subplots()

for name in ['Bitcoin', 'Ethereum', ]:
    ax.plot(df[df.Asset_Name==name].timestamp,df[df.Asset_Name==name].Close,label=name)

ax.set_xlabel("timestamp")
ax.set_ylabel("Close")
ax.legend(loc='best')

#a) this tells that the daily closing price of bitcoin is constantly getting way higher than of ethereum
#b) 
```

```python
#Question 4(a,b,c)
btc = df[df['Asset_Name'] == 'Bitcoin'].copy()
btc['SMA20'] = btc['Close'].rolling(20).mean()
btc.dropna(inplace=True)

btc['STD']= btc['Close'].rolling(20).std()
btc['LB']= btc['SMA20'] - btc['STD']*2
btc['UB']= btc['SMA20'] + btc['STD']*2
btc[['UB','LB','SMA20']].plot()
```
